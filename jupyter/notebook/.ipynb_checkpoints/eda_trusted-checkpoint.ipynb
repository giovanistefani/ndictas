```python
with open(json_meta,'rb') as json_meta:
     metadados = json.load(json_meta)
```

```python
def sel_no_int_cols(cols) :
    validatecols = []
    for col in cols :
        try :
            int (col)
        except :
            validatecols.append (col)
    return validatecols

def replacevalue(df, val, replace_val) :
    for col in df.columns :
        df[col] = df[col].apply (lambda x :re.sub (val, replace_val, str (x)))
    return df

def verify_type(df, dic) :
    l_type = []
    l_values = []
    l_validator = []
    for col in dic :
        try :
            int(col)
        except:
            try :
                if ((dic[col] == 'int' and type(int (df[col])) is int) or 
                    (dic[col] == 'float' and type(float (df[col])) is float) or 
                    (dic[col] == 'date' and (df[col]=='nan' or 
                                             datetime.strptime (df[col], '%d/%m/%Y'))) or 
                    dic[col]=='string') :
                    l_validator.append (True)
                    l_values.append('if')
                    l_type.append(dic[col])
                else :
                    l_validator.append (False)
                    l_values.append('else')
                    l_type.append(dic[col])
            except :
                l_validator.append (False)
                l_values.append('except')
                l_type.append(dic[col])
    if False in l_validator:
        return False , l_validator, l_values, l_type
    else:
        return True , l_validator, l_values, l_type        
    
    
def consult_params_table(tb_name, col) :
    for tab in metadados :
        if tab['tabela'] == tb_name :
            return [field[col] for field in tab['fields']]    
```

```python
blob_dir = 'prestador'
blob_file = 'PRESTADORES.pkl'

with open(f'{notebook_dir}/{blob_file}','wb') as data_from:
    data_from.write(cc_trusted.get_blob_client(f'{blob_dir}/{blob_file}').download_blob().readall())
    
df = pd.read_pickle(f'{notebook_dir}/{blob_file}')    
```

```python
tipo = consult_params_table(blob_dir, 'tipo') 
nome_banco = consult_params_table(blob_dir, 'nome_banco')
dict_tipo = dict (zip (nome_banco, tipo))
```

```python
df2 = df[sel_no_int_cols(df.columns)]
```

```python
df2 = replacevalue (df2, r'NULL|[\t]', '')
```

```python
df2 = df2.apply (lambda x :x.str.strip ().replace ('', None))
```

```python
df2[['validate','list_validate','l_values','l_types']] = df2.apply (lambda x :verify_type (x, dict_tipo), axis = 1, result_type='expand')
```
