```python
table3 = etl.fromcsv('datalake/RECEITA.csv', delimiter='|')
table3
```

```python
mappings = OrderedDict()
mappings['mes_competencia'] = 'mes_competencia'
mappings['fk_beneficiario'] = 'fk_beneficiario'
mappings['fk_empresa'] = 'fk_empresa'
mappings['dt_geracao_titulo'] = 'dt_geracao_titulo', {'NULL': 'aaaaa'}
mappings['dt_pgto'] = 'dt_pgto'
mappings['nr_centro_custo'] = 'nr_centro_custo'
mappings['descricao_centro_custo'] = 'descricao_centro_custo'
mappings['tipo_cobranca'] = 'tipo_cobranca'
mappings['vl_cobranca'] = 'vl_cobranca'
mappings['vl_pago'] = 'id'
mappings['tipo_cobranca_sub'] = 'tipo_cobranca_sub'
```

```python
table4 = etl.fieldmap(table3, mappings)
table4
```
